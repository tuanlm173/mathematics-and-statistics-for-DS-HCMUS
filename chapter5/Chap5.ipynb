## ANOVA

```python
df = pd.DataFrame({"Compact_cars": [643,655,702],
                  "Midsize_cars": [469,427,525],
                  "Full_size_cars": [484,456,402]})
```

```python
sns.boxplot(data=df)
plt.show()
```

```python
fvalue, pvalue = stats.f_oneway(df["Compact_cars"], df["Midsize_cars"], df["Full_size_cars"])
print(fvalue, pvalue)
```

```python
# pvalue significant
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
df_melt = pd.melt(df.reset_index(), id_vars=['index'], value_vars=["Compact_cars", "Midsize_cars", "Full_size_cars"])
```

```python
df_melt.columns = ["index", "cars", "value"]
```

```python
df_melt
```

```python
df
```

```python
model = ols("value ~ C(cars)", data=df_melt).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
```

```python
anova_table # Pvalue < 0.05, significant differences between type of cars
```

```python
# Perform multiple pairwise comparison
m_comp = pairwise_tukeyhsd(endog=df_melt["value"], groups=df_melt["cars"], alpha=0.05)
print(m_comp) # except last pair, otherwise, we rejects null hypothesis (mean are equal)
```

```python
# levene test for equal variances
w, pvalue = stats.levene(df["Compact_cars"], df["Midsize_cars"], df["Full_size_cars"])
print(w, pvalue) # pvalue > 0.05, false to reject null hypothesis, samples have same variance
```

```python
 # Shapiro test for sample from normal distribution
w, pvalue = stats.shapiro(model.resid)
print(w, pvalue) # pvalue > 0.05, sample from normal distribution
```
